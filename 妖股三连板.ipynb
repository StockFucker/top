```python
import pandas as pd
```

```python
data_df = pd.read_csv("top.csv",index_col=0,parse_dates=True,low_memory=False)
```

```python
data_df = data_df[(data_df.high0 == data_df.high_limit0) & (data_df.high1 == data_df.high_limit1) & (data_df.low0 < data_df.high_limit0) & (data_df.low1 < data_df.high_limit1)]
```

```python
data_df = data_df[data_df["date"] > "2012-01-01"]
```

```python
data_df = data_df[(data_df["date"] < "2015-01-01") | (data_df["date"] > "2015-06-01")]
```

```python
data_df = data_df.set_index(["stock","date"])
```
